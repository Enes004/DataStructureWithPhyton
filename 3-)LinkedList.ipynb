```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None


    def delete_circular(self, key):
        if self.head is None: return 
        temp = self.head
        
        while True:
            if temp.data == key: break
            temp = temp.next
            if temp == self.head: return 

        if temp.next == temp:
            self.head = self.tail = None
            return

        temp.prev.next = temp.next
        temp.next.prev = temp.prev

        if temp == self.head: self.head = temp.next
        if temp == self.tail: self.tail = temp.prev


    def deletefromhead(self):
        temp=self.head
        if self.head is None:
            return
        if temp.next == temp:
            self.head = self.tail=None
            return
            
        self.head = temp.next
        self.tail.next = self.head
        self.head.prev=self.tail

    def deletefromtail(self):
        temp=self.tail
        
        if self.head is None:
            return

        if temp.next == temp:
            self.head = self.tail = None
            return

        self.tail = temp.prev
        temp.prev.next = self.head
        self.head.prev = temp.prev
        
    def circular_search(self,key):
        temp = self.head
        if self.head is None:
            return

        while True:
            if temp.data == key:
                return temp
            temp=temp.next
            if temp == self.head:
                return False

        
    def DelByValue(self,key):
        if self.head is None:
            return
        
        node_to_del = self.circular_search(key)
        if node_to_del is None:
            return

        if node_to_del == self.head:
            self.deletefromhead()
            return
            
        if node_to_del == self.tail:
            self.deletefromtail()
            return
            
        node_to_del.prev.next=node_to_del.next
        node_to_del.next.prev=node_to_del.prev
        
```
