```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def Enqueue(self,data): #sona eklemek
        NewNode = Node(data)
        if self.front is None:
            self.front = self.rear = NewNode
            return
        self.rear.next = NewNode
        self.rear = NewNode

    def Dequeue(self): #bastan sil
        if self.front is None:
            return
        deq_data = self.front.data
        if self.front.next is None:
            self.front = self.rear = None
            return deq_data
        
        self.front = self.front.next
        return deq_data        
            
```

```python
q = Queue()
q.Enqueue(10)
q.Enqueue(20)
q.Enqueue(30)
```
