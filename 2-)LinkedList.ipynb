```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    #O(1)
    def insert_2_end(self,data):
        NewNode = Node(data)
        if self.head is None:
            self.head = self.tail = NewNode
            return

        self.tail.next = NewNode
        NewNode.prev = self.tail
        self.tail = NewNode    
    
    def insert2head(self,data): 
        NewNode = Node(data)
        if self.head is None:
           self.head=self.tail=NewNode
           return

        NewNode.next = self.head
        self.head.prev = NewNode
        self.head = NewNode

    def display(self):
        temp = self.head
        while temp :
            print(f"{temp.data} <->",end=" ")
            temp=temp.next
        print("None")
            
    def display_reverse(self):
        temp = self.tail
        while temp is not None:
            print(f"{temp.data} <->",end=" ")
            temp= temp.prev
        print("None")

    def display_recursive(self,Node):
        if Node is None:
            print("None")
            return
        print(f"{Node.data} <->",end=" ")
        self.display_recursive(Node.next)

    def delete_by_value(self,key):
        temp=self.head
        # if list is empty
        if self.head is None:
            return
            #if list consist of just self.head
        if self.head.next is None and self.head.data==key:
                self.head = self.tail = None
                return
            
        if self.head.data == key:
            self.head = temp.next
            temp.next=None
            self.head.prev = None
            return

        if self.tail.data != key:
            while temp is not None and temp.data != key:
                temp=temp.next
                if temp is None:
                    return
            pre = temp.prev
            pre.next = temp.next
            temp.next.prev = pre
            temp = None
            return

        if self.tail.data == key:
            erase = self.tail
            self.tail = erase.prev
            self.tail.next = None
            erase = None
            return

        return

    def delAnOde(self,key):
        temp = self.head

            #bring temp to key value
        while temp and temp.data != key:
                temp = temp.next
            # if temp is none return
        if temp is None:     
             return

          #Checking left and right
        if temp ==self.head:
             self.head = temp.next
        else:
             temp.prev.next=temp.next

        if temp == self.tail:
             self.tail= temp.prev
        else:
            temp.next.prev=temp.prev
```
