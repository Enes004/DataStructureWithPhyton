```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None


    #O(N)
    def insert2end(self,data):
        NewNode = Node(data)
        if self.head is None:
            self.head = NewNode
            return

        last = self.head
        while (last.next is not None):
            last = last.next

        last.next = NewNode
        NewNode.prev = last

    #O(1)
    def insert_2_end(self,data):
        NewNode = Node(data)
        if self.head is None:
            self.head = self.tail = NewNode
            return

        self.tail.next = NewNode
        NewNode.prev = self.tail
            
    
    def insert2head(self,data): 
        NewNode = Node(data)
        if self.head is None:
           self.head=self.tail=NewNode
           return

        NewNode.next = self.head
        self.head.prev = NewNode
        self.head = NewNode
```
