```python
# STRUCTURE OF THE LINEAR / ONE-WAY LINKED LIST

class Node:
    def __init__(self,data):             # Equilavent to Struct Node in C.
        self.data = None                 # Represents a single Node in the list.
        self.next = None


class LinkedList:                      
    def __init__(self,data):             # Holds the head pointer.
        self.head = None                 # Manage the Linked List operations.


def Insert2End(self,value):
    NewNode = Node(value)  # Create a node
    
    #if list is empty just add the node as head.
    if self.head is None:
        NewNode = self.head

    #else:
    last = self.head
    while last.next != None:
        last = last.next

    last.next = NewNode
    
        
def Insert2head(self,value):
    NewNode = Node(value)

    if self.head == None:
        self.head = NewNode

    temp = self.head
    NewNode.next = temp
    self.head = NewNode

def DeleteByValue(self,key):
    # Control if list is null
    if self.head == None:
        return

    temp = self.head
    
    # Control if HEAD is target
    if (temp is not None) and (temp.data == key):
        self.head = temp.next
        temp = None
    
    prev = None
    # Search until find the target
    while (temp is not none) and (temp.value != key):
          prev = temp
          temp = temp.next
    # if couldn't find return
    if temp is None:
        return
    # if found the key 
    prev.next=temp.next
    temp = None

    

def PrintList(self):
    temp = self.head
    while temp is not None:
        print(f"{temp.data} ->",end=" ")
        temp = temp.next
    print("None")

    
```
