```python
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self,data):
        if self.root is None:
            self.root = Node(data)
        else:
            self.insert_recursive(self.root,data)

    def insert_recursive(self,curr,data):
        if data < curr.data:
            if curr.left is None:
                curr.left = Node(data)
            else:
                self.insert_recursive(curr.left,data)

        elif data > curr.data:
            if curr.right is None:
                curr.right = Node(data)
            else:
                self.insert_recursive(curr.right,data)

    def search(self,data):
        return self.recursive_search(self.root,data)

    def recursive_search(self,curr,data):
        if curr is None:
            return False

        elif curr.data == data:
            return True

        elif data < curr.data:
            return self.recursive_search(curr.left,data)

        elif data > curr.data:
            return self.recursive_search(curr.right,data)

    def print_tree(self):
        self.inorder_print_tree(self.root)

    def inorder_print_tree(self,curr):
        if curr is not None:
            self.inorder_print_tree(curr.left)
            print(curr.data , end=" ")
            self.inorder_print_tree(curr.right)
```

```python
bst = BST()
```

```python
print("--- Sayılar Ekleniyor ---")
# Sayıları şu sırayla ekliyoruz: 50, 30, 20, 40, 70, 60, 80
bst.insert(50) # Kök
bst.insert(30)
bst.insert(20)
bst.insert(40)
bst.insert(70)
bst.insert(60)
bst.insert(80)

# 1. TEST: Sıralı Yazdırma
print("Ağacın Sıralı Hali (Beklenen: 20 30 40 50 60 70 80):")
bst.print_tree() 

# 2. TEST: Arama Yapma
aranan = 60
print(f"\n{aranan} sayısı ağaçta var mı?: {bst.search(aranan)}")

aranan_yok = 99
print(f"{aranan_yok} sayısı ağaçta var mı?: {bst.search(aranan_yok)}")
```
