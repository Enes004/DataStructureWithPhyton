```python
# Adj_list
#  {
#    "A" : {"B":20, "C":30}     A , B and C are vertexes , A=Destination , B=Target / A=Destination C=Target , 20 and 30 are weight
#    "k" : {"C":22, "H":10}
#    "L" : {"Z":24, "R":37}# "M" : {"T":28, "Q":34}
#  }
```

```python
class Graph:
    def __init__(self):
        self.adj_list = {}

    def add_vertex(self,vertex):
        if vertex not in self.adj_list:
            self.adj_list[vertex] = {}

    def add_edge(self,source,destination,weight):
        self.add_vertex(source)
        self.add_vertex(destination)

        self.adj_list[source][destination] = weight
    
            
    def print_graph(self):
        for source,neighbours in self.adj_list.items():
            for destination,weight in neighbours.items():
                print(f"Vertex={source} --- {weight} --- Vertex={destination}")


    def remove_edge(self,source,destination):
        if source in self.adj_list and destination in self.adj_list[source]:
            del self.adj_list[source][destination]
            print(f"Removed {source}---{destination}")
        else:
            print("No such a edge")


    def get_weight(self,source,destination):
        if source in self.adj_list and destination in self.adj_list[source]:
            return self.adj_list[source][destination]
        else:
            return None

    def get_neighbors(self,source):
        if source in self.adj_list:
            neighbours = list(self.adj_list[source].keys())
            return neighbours
        else:
            return None
```

```python
g = Graph()
g.add_edge("İstanbul", "Ankara", 450)
g.add_edge("İstanbul", "İzmir", 480)
g.add_edge("Ankara", "Konya", 260)

print("--- İlk Durum ---")
g.print_graph()

print("\n--- Sorgular ---")
print(f"İstanbul'un komşuları: {g.get_neighbors('İstanbul')}")
print(f"İstanbul-Ankara arası mesafe: {g.get_weight('İstanbul', 'Ankara')} km")

print("\n--- Silme İşlemi ---")
g.remove_edge("İstanbul", "İzmir") # İzmir yolunu iptal et
g.print_graph()
```
