```python
def ValidParantheses(text):
    stack = []
    matching = {
             ')':'(',
             '}':'{',
             ']':'[' }
    #)
    #}
    #]
    #[ 3
    #{ 2
    #( 1

    
    for ch in text:
        if ch in matching:
            if not stack or (matching[ch] != stack[-1]):
                return False
            stack.pop()

        else:
            stack.append(ch)

    return not stack
            
        
```
