```python
# STRUCTURE OF THE LINEAR / ONE-WAY LINKED LIST

class Node:
    def __init__(self,data):             # Equilavent to Struct Node in C.
        self.data = data                 # Represents a single Node in the list.
        self.next = None


class LinkedList:                      
    def __init__(self):                  # Holds the head pointer.
        self.head = None                 # Manage the Linked List operations.


    def Insert2End(self,data):
        NewNode = Node(data)  # Create a node
    
        #if list is empty just add the node as head.
        if self.head is None:
             self.head=NewNode
             return

        #else:
        last = self.head
        while last.next != None:
            last = last.next

        last.next = NewNode
    
        
    def Insert2head(self,data):
        NewNode = Node(data)

        if self.head == None:
            self.head = NewNode
            return
        
        temp = self.head
        NewNode.next = temp
        self.head = NewNode

    def DeleteByValue(self,key):
        # Control if list is null
        if self.head == None:
            return

        temp = self.head
    
        # Control if HEAD is target
        if (temp is not None) and (temp.data == key):
            self.head = temp.next
            temp = None
            return
    
        prev = None
        # Search until find the target
        while (temp is not None) and (temp.data != key):
              prev = temp
              temp = temp.next
        # if couldn't find return
        if temp is None:
            return
        # if found the key 
        prev.next=temp.next
        temp = None
    
    

    def PrintList(self):
        temp = self.head
        while temp is not None:
            print(f"{temp.data} ->",end=" ")
            temp = temp.next
        print("None")

    
```

```python
if __name__ == "__main__":
    my_list = LinkedList()

    my_list.Insert2End(1)
    my_list.Insert2End(2)
    my_list.Insert2End(3)
    my_list.Insert2End(4)
    my_list.PrintList()
    
```

```python
my_list.Insert2head(5)
my_list.Insert2head(6)
my_list.PrintList()
```

```python
my_list.DeleteByValue(5)
my_list.PrintList()
```
