```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None


    def push(self,data):
        NewNode = Node(data)
        if self.top is None:
            self.top = NewNode        #we can delete if condition here,it doesnt need
            return                  
        NewNode.next = self.top
        self.top = NewNode

    def pop(self):
        if self.top is None:
            return
        peek = self.top.data
        temp = self.top.next
        self.top.next = None
        self.top = temp
        return peek
```

```python
s = Stack()
s.push(10)
s.push(20)
s.push(30)
```

```python
print(s.pop())
```
